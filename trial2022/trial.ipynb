```python
def clump(a):
    return [a[s] for s in np.ma.clump_unmasked(np.ma.masked_invalid(a))]
```

```python
clumps= clump(list(df[0]))
```

```python
sumclumps = [sum(i) for i in clumps]
```

```python
sumclumps.sort()
```
